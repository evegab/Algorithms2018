## Soluci√≥n punto 2 en Python usando numpy 

```python
import numpy as np
import matplotlib.pyplot as plt
import time

f0 = np.int_( 0 )
f1 = np.int_( 1)

n = np.int_( 42 )
t0 = 0
tf = 0
t1 = 0
listTime = []

for i in range ( 0, int(n + 1) ):
	t0 = time.clock()
	#print "t0 = ", t0
	f1 = f0 + f1
	f0 = f1 - f0
	t1 = time.clock()
	tf = t1 - t0
	listTime.append( tf )
	#print "t1 = ", t1
	#print "tf = ", tf
	#print f1
	
sizeArray = range ( n + 1 ) # lista del 0 to 11 
casoPromedio = listTime

fig = plt.figure()
axes = fig.add_subplot( 111 )
plt.bar( sizeArray, casoPromedio )
plt.ylabel('Tiempo' )
plt.xlabel('Tamano n')
plt.title( "n vs time" )
plt.show()
```
